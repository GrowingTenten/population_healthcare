```python
import pandas as pd
import numpy as np
import seaborn as sns
from glob import glob
import matplotlib as mpl
import matplotlib.pylab as plt
import folium
import json
```

```python
df_oc = pd.read_csv("미드프로젝트/data/df_oc_depart_rate.csv", encoding="utf8")
```

```python
df_oc['시군구명'] = df_oc['시군구명'].str.replace('고양','고양시')
df_oc['시군구명'] = df_oc['시군구명'].str.replace('성남','성남시')
df_oc['시군구명'] = df_oc['시군구명'].str.replace('수원','수원시')
df_oc['시군구명'] = df_oc['시군구명'].str.replace('안산','안산시')
df_oc['시군구명'] = df_oc['시군구명'].str.replace('안양','안양시')
df_oc['시군구명'] = df_oc['시군구명'].str.replace('용인','용인시')
```

```python


# 서울
df_oc_seoul = df_oc[(df_oc["시도명"]=="서울특별시")]


df_oc_seoul = df_oc_seoul.groupby(["시군구명","요양종별"])[["요양종별의료기관수"]].sum().reset_index()
df_oc_seoul_beauty = df_oc_seoul.loc[(df_oc_seoul["요양종별"]=="성형외과") | (df_oc_seoul["요양종별"]=="피부과")]
df_oc_seoul_other = df_oc_seoul.loc[(df_oc_seoul["요양종별"]=="내과") | (df_oc_seoul["요양종별"]=="산부인과") | (df_oc_seoul["요양종별"]=="외과") | (df_oc_seoul["요양종별"]=="안과") | (df_oc_seoul["요양종별"]=="치과")]

df_oc_seoul_beauty = df_oc_seoul_beauty.groupby(["시군구명"])[["요양종별의료기관수"]].sum()
df_oc_seoul_beauty = df_oc_seoul_beauty.rename(columns = {"요양종별의료기관수" : "서울미용"})
df_oc_seoul_other = df_oc_seoul_other.groupby(["시군구명"])[["요양종별의료기관수"]].sum()
df_oc_seoul_other = df_oc_seoul_other.rename(columns = {"요양종별의료기관수" : "서울비미용"})


df_oc_seoul_total = pd.concat([df_oc_seoul_beauty, df_oc_seoul_other],axis=1)

df_oc_seoul_total = df_oc_seoul_total.fillna(0)
df_oc_seoul_total = df_oc_seoul_total.reset_index()

df_oc_seoul_total = df_oc_seoul_total.assign(미용목적_병원_비율=lambda x:(x.서울미용 / (x.서울미용+x.서울비미용))*100)
df_oc_seoul_total = df_oc_seoul_total.assign(비미용목적_병원_비율=lambda x:(x.서울비미용 / (x.서울미용+x.서울비미용))*100)



# 경기도 

df_oc_gg = df_oc[(df_oc["시도명"]=="경기도")]

df_oc_gg = df_oc_gg.groupby(["시군구명","요양종별"])[["요양종별의료기관수"]].sum().reset_index()
df_oc_gg_beauty = df_oc_gg.loc[(df_oc_gg["요양종별"]=="성형외과") | (df_oc_gg["요양종별"]=="피부과")]
df_oc_gg_other = df_oc_gg.loc[(df_oc_gg["요양종별"]=="내과") | (df_oc_gg["요양종별"]=="산부인과") | (df_oc_gg["요양종별"]=="외과") | (df_oc_gg["요양종별"]=="안과") | (df_oc_gg["요양종별"]=="치과")]


df_oc_gg_beauty = df_oc_gg_beauty.groupby(["시군구명"])[["요양종별의료기관수"]].sum()
df_oc_gg_other = df_oc_gg_other.groupby(["시군구명"])[["요양종별의료기관수"]].sum()
df_oc_gg_beauty = df_oc_gg_beauty.rename(columns = {"요양종별의료기관수" : "경기도미용"})
df_oc_gg_other = df_oc_gg_other.rename(columns = {"요양종별의료기관수" : "경기도비미용"})

df_oc_gg_total = pd.concat([df_oc_gg_beauty, df_oc_gg_other], axis=1)

df_oc_gg_total_final = df_oc_gg_total.fillna(0)
df_oc_gg_total_final.reset_index()

df_oc_gg_total_final = df_oc_gg_total_final.assign(미용목적_병원_비율=lambda x:(x.경기도미용 / (x.경기도미용+x.경기도비미용))*100)
df_oc_gg_total_final = df_oc_gg_total_final.assign(비미용목적_병원_비율=lambda x:(x.경기도비미용 / (x.경기도미용+x.경기도비미용))*100)
```
